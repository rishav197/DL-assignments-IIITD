Normalizing the X_train and X_test datasets within the range of [0, 1]

```python
X_train = X_train/255.0
X_valid = X_valid/255.0
X_test = X_test/255.0
```

#Setup-1

```python
cnn1 = models.Sequential([
    layers.Conv2D(filters=10, kernel_size=(5, 5), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D(),

    layers.Conv2D(filters=20, kernel_size=(2, 2), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D(),

    #mlp(Dense Network)
    layers.Flatten(),
    layers.Dense(50, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
cnn1.compile(optimizer='adam', 
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
cnn1_train_history = cnn1.fit(X_train, y_train, epochs=10)
```

```python
cnn1_valid_history = cnn1.fit(X_valid, y_valid, epochs=10)
```

```python
train_loss, train_accuracy = cnn1_train_history.history['loss'], cnn1_train_history.history['accuracy']
valid_loss, valid_accuracy = cnn1_valid_history.history['loss'], cnn1_valid_history.history['accuracy']
```

```python
epoch_arr = [i for i in range(1, len(train_accuracy)+1)]
accuracy_arr = [ele*100 for ele in train_accuracy]
```

```python
print(len(train_loss), train_loss)
print(len(train_accuracy), train_accuracy)
print(len(epoch_arr), epoch_arr)
```

```python
print(len(valid_loss), valid_loss)
print(len(valid_accuracy), valid_accuracy)
```

4. Plots for the above setup

```python
# plot for training accuracy and validation accuracy
plt.plot(epoch_arr, train_accuracy)
plt.scatter(epoch_arr, train_accuracy)
plt.title("Plot: Train Accuracy vs Epochs")
plt.ylabel("Train Accuracy")   
plt.xlabel("Epochs")
plt.savefig("train_accu_epochs.png", bbox_inches="tight", pad_inches=2, transparent=False)
plt.show()


plt.plot(epoch_arr, valid_accuracy)
plt.scatter(epoch_arr, valid_accuracy)
plt.title("Plot: Validation Accuracy vs Epochs")
plt.ylabel("Validation Accuracy")   
plt.xlabel("Epochs")
plt.savefig("valid_accu_epochs.png", bbox_inches="tight", pad_inches=2, transparent=False)
plt.show()
```
